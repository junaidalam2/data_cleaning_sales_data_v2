```python
import pandas as pd
```

```python
df = pd.read_csv('../data/retail_store_sales.csv')
df
```

```python
print(df.describe())
```

```python
df = df.fillna('')
df
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
df['Transaction ID'] = df['Transaction ID'].str.strip()
transID_format_mask = (
    (df['Transaction ID'].str[:4] != 'TXN_') | 
    (df['Transaction ID'].str.len() != 11) | 
    (~df['Transaction ID'].str[4:].str.isnumeric())
    )

filtered_transID_format = [item for item in transID_format_mask if item]
filtered_transID_format
```

All cells in *Transaction ID* meet required format.

```python
df['Customer ID'] = df['Customer ID'].str.strip()
customerID_format_mask = (
    (df['Customer ID'].str[:5] != 'CUST_') | 
    (df['Customer ID'].str.len() != 7) | 
    (~df['Customer ID'].str[5:].str.isnumeric())
    )

filtered_customerID_format = [(idx, item) for idx, item in customerID_format_mask.items() if item]
filtered_customerID_format
```

All cells in *Customer ID* meet required format.

```python
df['Category'] = df['Category'].str.strip()
unique_categories = df['Category'].unique()
unique_categories
```

No information to clean in *Category*. 

```python
df['Item'] = df['Item'].str.strip()
df
```

```python
unique_item = df['Item'].unique()
unique_item
```

Not possible to reconstruct missing data in *Item* column. Item codes have subtypes and do not map directly to the *Category* field. For instance, 'Item_11_FOOD', 'Item_23_FOOD', 'Item_3_FOOD' are all mapped to 'Food' within the field *Category*.

```python
length_before = df.shape[0]
df = df[df['Item'] != '']
length_after = df.shape[0] 
df
```

```python
num_removed_rows = length_before - length_after
print('Number of Rows Removed: ', num_removed_rows, '| Percentage of Rows Removed: ', num_removed_rows / length_before)
```

Assumption that *Item* required for analysis of dataset. Removal of < 10% of records determined to be acceptable.

```python
unique_price = df['Price Per Unit'].unique()
unique_price
```

```python
unique_quantity = df['Quantity'].unique()
unique_quantity
```

No information to clean in *Quantity*.

```python
df['Total Spent Recalc'] = df['Price Per Unit'] * df['Quantity']
df['Total Spent Check'] = df['Total Spent'] - df['Total Spent Recalc']
df
```

```python
total_spent_check_mask = (
    (df['Total Spent Check'] != 0)
    )

total_spent_check_mask

filtered_total_spent_check = [(idx, item) for idx, item in total_spent_check_mask.items() if item]
print(len(filtered_total_spent_check))
print(filtered_total_spent_check)
```

No information to clean in *Total Spent*.

```python
df = df.drop(columns =['Total Spent Recalc', 'Total Spent Check'])
df
```

```python
unique_payment_method = df['Payment Method'].unique()
unique_payment_method
```

```python
unique_location = df['Location'].unique()
unique_location
```

```python
cash_payment_online_mask = (
    (df['Payment Method'] == 'Cash') &
    (df['Location'] == 'Online')    
    )

cash_payment_In_Store_mask = (
    (df['Payment Method'] == 'Cash') &
    (df['Location'] == 'In-store')    
    )

filtered_cash_payment_online = [(idx, item) for idx, item in cash_payment_online_mask.items() if item]
filtered_cash_payment_in_store = [(idx, item) for idx, item in cash_payment_In_Store_mask.items() if item]

print('No. Cash Payments Online: ', len(filtered_cash_payment_online), '| No. Cash Payments In-Shore: ', len(filtered_cash_payment_in_store))
```

Likely a good explanation for classification of cash payments online when considering combination of data within columns *Payment Method* and *Location* given that ~50% of cash sales are on-line. Obtain further information from providers of information.

```python
df[['Transaction Date Year', 'Transaction Date Month', 'Transaction Date Day']] = df['Transaction Date'].str.split('-', expand=True)
df['Transaction Date Year'] = pd.to_numeric(df['Transaction Date Year'], errors = 'coerce')
df['Transaction Date Month'] = pd.to_numeric(df['Transaction Date Month'], errors = 'coerce')
df['Transaction Date Day'] = pd.to_numeric(df['Transaction Date Day'], errors = 'coerce')
df
```

```python
unique_tn_date_year = df['Transaction Date Year'].unique()
unique_tn_date_month = df['Transaction Date Month'].unique()
unique_tn_date_day = df['Transaction Date Day'].unique()


print("Year:", unique_tn_date_year)
print("Month: ", unique_tn_date_month)
print("Day: ", unique_tn_date_day)
```

```python
df = df.drop(columns = ['Transaction Date Year', 'Transaction Date Month', 'Transaction Date Day'])
```
