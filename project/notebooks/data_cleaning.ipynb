```python
import pandas as pd
```

```python
df = pd.read_csv('../data/retail_store_sales.csv')
df
```

```python
print(df.describe())
```

```python
df = df.fillna('')
df
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
df['Transaction ID'] = df['Transaction ID'].str.strip()
transID_format_mask = (
    (df['Transaction ID'].str[:4] != 'TXN_') | 
    (df['Transaction ID'].str.len() != 11) | 
    (~df['Transaction ID'].str[4:].str.isnumeric())
    )

filtered_transID_format = [item for item in transID_format_mask if item]
filtered_transID_format
```

All cells in *Transaction ID* meet required format.

```python
df['Customer ID'] = df['Customer ID'].str.strip()
customerID_format_mask = (
    (df['Customer ID'].str[:5] != 'CUST_') | 
    (df['Customer ID'].str.len() != 7) | 
    (~df['Customer ID'].str[5:].str.isnumeric())
    )

filtered_customerID_format = [(idx, item) for idx, item in customerID_format_mask.items() if item]
filtered_customerID_format
```

All cells in *Customer ID* meet required format.

```python
df['Category'] = df['Category'].str.strip()
unique_categories = df['Category'].unique()
unique_categories
```

No information to clean in *Category*. 

```python
df['Item'] = df['Item'].str.strip()
df
```

```python
unique_item = df['Item'].unique()
unique_item
```

Not possible to reconstruct missing data in *Item* column. Item codes have subtypes and do not map directly to the *Category* field. For instance, 'Item_11_FOOD', 'Item_23_FOOD', 'Item_3_FOOD' are all mapped to 'Food' within the field *Category*.

```python
length_before = df.shape[0]
df = df[df['Item'] != '']
length_after = df.shape[0] 
df
```

```python
num_removed_rows = length_before - length_after
print('Number of Rows Removed: ', num_removed_rows, '| Percentage of Rows Removed: ', num_removed_rows / length_before)
```
